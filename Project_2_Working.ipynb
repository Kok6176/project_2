```python
# Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#Go to site
url ="https://www.lifehack.org/articles/communication/50-new-years-resolution-ideas-and-how-achieve-each-them.html"
browser.visit(url)
```

```python
# Create a Beautiful Soup object
html=browser.html
soup_obj = soup(html,'html.parser')
```

```python
# Extract all the headers 
goals = soup_obj.find_all('div', class_='bialty-container')
```

```python
i=1
title_new=[]
for result in goals:
    for title in result.find_all('h2'):
        repl_str = str(i)+". "
        value_new = title.text.replace(repl_str,'')
        i = i + 1
        print(value_new)
        title_new.append(value_new)
```

```python
# pd.DataFrame({'email':sf.index, 'list':sf.values})

# df = pd.DataFrame({'id':title_new.index, 'resolution_topics':title_new.values})
df = pd.DataFrame(title_new)
df1 = df.rename(columns={0: 'resolution_topics'})
df1
```

```python
def match_resol_to_key(resol):
    for key, value in resol_dict.items():
        if resol in value:
            return key


df2['mod_resol'] = df2['resolution_topics'].apply(match_resol_to_key)
df2
```

```python
new_yr_res = pd.merge(df1, df2, left_on='resolution_topics', right_on ='mod_resol', how='left')
new_yr_res
```

```python
from sqlalchemy import create_engine
from pg import pg_key
import psycopg2

# building a connection string to connect to our postgres db

dbengine = "postgresql://postgres:" + pg_key +"@localhost:5432/new_resolution_db"
engine = create_engine(dbengine)
```

```python
# load merged dataframes into database(POSTGRES)
new_yr_res.to_sql(name='newYearResolution', con=engine, if_exists='replace', index=False)
```
